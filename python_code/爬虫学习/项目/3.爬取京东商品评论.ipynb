https://sclub.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98vv516&productId=6946635&score=0&sortType=5&page=2&pageSize=10&isShadowSku=0&rid=0&fold=1  

https://sclub.jd.com/comment/productPageComments.action?callback=fetchJSON_comment98vv516&productId=6946635&score=0&sortType=5&page=3&pageSize=10&isShadowSku=0&rid=0&fold=1

分析以上的请求，得到规律

```python
import requests
from urllib.parse import urlencode

base_url = 'https://sclub.jd.com/comment/productPageComments.action?'
page = 3
params = {
    'callback':'fetchJSON comment98vv516',
    'productId':'6946635',
    'score':'0',
    'sortType':'5',
    'page':page,
    'pageSize':'10',
    'isShadowSku':'0',
    'rid':'0',
    'fold':'1'
}

url = base_url + urlencode(params)
response = requests.get(url)
print(response.status_code)
```

```python
# 看一下文本内容
print(response.text)
```

```python
# 剔除多余的部分，得到字典，便于后面处理
import re
result = re.match('.*?\\((.*)\\);', response.text)
```

```python
print(result.group(1))
```

```python
print(type(result.group(1)))
```

```python
# 转换为json格式
import json
data = json.loads(result.group(1))
```

```python
print(data)
```

```python
print(type(data))
```

```python
print(data.keys())
```

```python
# 得到评论
coms = data.get('comments')
```

```python
print(type(coms))
```

```python
print(len(coms))
```

```python
# 遍历输出每个评论的ID和评论内容
for item in coms:
    print(item.get('id'))
    print(item.get('content').strip())
```
