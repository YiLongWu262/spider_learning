#### 爬取买眼电影Top100榜单

**先抓取一个页面分析一下我们所要分析的数的特点**

```python
import requests
from bs4 import BeautifulSoup

base_url = 'http://maoyan.com/board/4?offset='
offset = 10
url = base_url + str(offset)

headers = {
    'User-Agent':'Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/55.0.2883.87 Safari/537.36'
}

try:
    response = requests.get(url, headers = headers)
    print(response.status_code)
    
except:
    print('ERRROR')
```

```python
soup = BeautifulSoup(response.text, "lxml")
```

```python
print(soup.prettify())
```

**找到电影的节点**

```python
print(soup.dd)
```

**找到电影的排名**

```python
print(soup.dd.i.get_text())
```

**找到电影名称**

```python
print(soup.dd.p.a['title'])
```

**找到电影的海报的链接**

```python
print(soup.dd.a.find_all('img')[1]['data-src'])
```

**找到所有的主演**

```python
print(soup.dd.find_all('p')[1].get_text().split())
print(type(soup.dd.find_all('p')[1].get_text().split()))
print(len(soup.dd.find_all('p')[1].get_text().split()[0]))
print(type(soup.dd.find_all('p')[1].get_text().split()[0]))
print(soup.dd.find_all('p')[1].get_text().split()[0][3:])
```

**找到上映时间和地点**

```python
print(soup.dd.find_all('p')[2].get_text().split()[0][5:])
```
