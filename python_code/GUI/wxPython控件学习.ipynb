wx.Panel

```python
import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        #创建panel
        self.panel = wx.Panel(parent=self)
        self.button1 = wx.Button(self.panel, label="Button1")
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.Panel")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
```

wx.StaticText  
A static text control displays one or more lines of read-only text

```python
import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
        
    def InitUI(self):
        #垂直
        self.box = wx.BoxSizer(wx.VERTICAL)
        
        #创建panel
        self.panel = wx.Panel(self)
        self.font = wx.Font(18, wx.ROMAN, wx.ITALIC, wx.NORMAL)
        
        #设置label1
        self.txt1 = "Python1"
        self.label1 = wx.StaticText(self.panel, style=wx.ALIGN_CENTER)
        self.label1.SetFont(self.font)
        self.label1.SetLabel(self.txt1)
        
        #设置label2
        self.txt2 = "Python2"
        self.label2 = wx.StaticText(self.panel, style=wx.ALIGN_CENTER)
        self.label2.SetFont(self.font)
        self.label2.SetLabel(self.txt2)
        
        #设置label3
        self.txt3 = "Python3"
        self.label3 = wx.StaticText(self.panel, style=wx.ALIGN_CENTER)
        self.label3.SetFont(self.font)
        self.label3.SetLabel(self.txt3)
        
        #将所有label加入到box中
        self.box.Add(self.label1)
        self.box.Add(self.label2)
        self.box.Add(self.label3)
        
        self.panel.SetSizer(self.box)
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.StaticText")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
```

二维网络布局

```python

import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        
        #初始化UI
        self.InitUI()
        
    def InitUI(self):
        #创建panel
        p = wx.Panel(self)
        #二维网络布局
        gs = wx.GridSizer(4, 4, 5, 5)
        
        #添加Button
        for i in range(1,17): 
             btn = "Btn"+str(i) 
             gs.Add(wx.Button(p,label = btn),0,wx.EXPAND) 

        p.SetSizer(gs)
                
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="Grid Demo")#创建一个顶层窗口
        self.frame.Centre()#居中
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
```

wx.TextCtrl

```python

import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        #创建panel
        panel = wx.Panel(self) 
        vbox = wx.BoxSizer(wx.VERTICAL) 

        #box1
        hbox1 = wx.BoxSizer(wx.HORIZONTAL) 
        l1 = wx.StaticText(panel, -1, "文本域") 
        hbox1.Add(l1, 1, wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        #t1
        self.t1 = wx.TextCtrl(panel) 
        hbox1.Add(self.t1,1,wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        self.t1.Bind(wx.EVT_TEXT,self.OnKeyTyped)
        #add
        vbox.Add(hbox1) 

        #box2
        hbox2 = wx.BoxSizer(wx.HORIZONTAL)
        l2 = wx.StaticText(panel, -1, "密码文本") 
        hbox2.Add(l2, 1, wx.ALIGN_LEFT|wx.ALL,5)
        #t2
        self.t2 = wx.TextCtrl(panel,style = wx.TE_PASSWORD)
        self.t2.SetMaxLength(5) 
        hbox2.Add(self.t2,1,wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        #add
        vbox.Add(hbox2)
        self.t2.Bind(wx.EVT_TEXT_MAXLEN,self.OnMaxLen)

        #box3
        hbox3 = wx.BoxSizer(wx.HORIZONTAL) 
        l3 = wx.StaticText(panel, -1, "多行文本") 
        hbox3.Add(l3,1, wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        #t3
        self.t3 = wx.TextCtrl(panel,size = (200,100),style = wx.TE_MULTILINE) 
        hbox3.Add(self.t3,1,wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        #add
        vbox.Add(hbox3)
        self.t3.Bind(wx.EVT_TEXT_ENTER,self.OnEnterPressed)  

        #box4
        hbox4 = wx.BoxSizer(wx.HORIZONTAL) 
        l4 = wx.StaticText(panel, -1, "只读取文本") 
        hbox4.Add(l4, 1, wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5)
        #t4
        self.t4 = wx.TextCtrl(panel, value = "只读文本",style = wx.TE_READONLY|wx.TE_CENTER) 
        hbox4.Add(self.t4,1,wx.EXPAND|wx.ALIGN_LEFT|wx.ALL,5) 
        
        vbox.Add(hbox4)
        panel.SetSizer(vbox) 

        self.Centre()
        self.Show()
        self.Fit()
        
    def OnKeyTyped(self, event): 
        print(event.GetString())

    def OnEnterPressed(self,event): 
        print("Enter pressed")

    def OnMaxLen(self,event): 
        print("Maximum length reached")
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.TextCtrl实例")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()

```

wx.RadioButton & wx.RadioBox

```python

import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(300, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        #创建panel
        pnl = wx.Panel(self)

        self.rb1 = wx.RadioButton(pnl,11, label = 'Value A',
                                pos = (10,10), style = wx.RB_GROUP) 
        self.rb2 = wx.RadioButton(pnl,22, label = 'Value B',pos = (10,40)) 
        self.rb3 = wx.RadioButton(pnl,33, label = 'Value C',pos = (10,70))
        #bind
        self.Bind(wx.EVT_RADIOBUTTON, self.OnRadiogroup)

        lblList = ['Value X', 'Value Y', 'Value Z'] 

        self.rbox = wx.RadioBox(pnl, label = 'RadioBox', pos = (80,10), choices = lblList,
        majorDimension = 1, style = wx.RA_SPECIFY_ROWS) 
        self.rbox.Bind(wx.EVT_RADIOBOX,self.onRadioBox) 

        self.Centre() 
        self.Show(True)
        
    def OnRadiogroup(self, e): 
        rb = e.GetEventObject() 
        print(rb.GetLabel()+' is clicked from Radio Group')

    def onRadioBox(self,e): 
        print(self.rbox.GetStringSelection()+' is clicked from Radio Box')
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="RadioButton & RadioBox")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()

```

wx.CheckBox 复选框

```python
import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        pnl = wx.Panel(self) 
        self.cb1 = wx.CheckBox(pnl, label = 'Value A',pos = (10,10)) 
        self.cb2 = wx.CheckBox(pnl, label = 'Value B',pos = (10,40)) 
        self.cb3 = wx.CheckBox(pnl, label = 'Value C',pos = (10,70)) 
        self.Bind(wx.EVT_CHECKBOX,self.onChecked) 
        self.Centre() 
        self.Show(True)
        
    def onChecked(self, e): 
        cb = e.GetEventObject() 
        print(cb.GetLabel()+' is clicked'+' '+str(cb.GetValue()))
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.Panel")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
    
    
```

wx.ComboBox对象提供从项目选择列表。它可以被配置为一个下拉列表或永久性的显示。

```python

import wx

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 400)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        #创建panel
        panel = wx.Panel(self)
        
        box = wx.BoxSizer(wx.VERTICAL)
        
        self.label = wx.StaticText(panel,label = "Your choice:" ,style = wx.ALIGN_CENTRE) 
        box.Add(self.label, 0 , wx.EXPAND |wx.ALIGN_CENTER_HORIZONTAL |wx.ALL, 20)
        
        cblbl = wx.StaticText(panel,label = "Combo box",style = wx.ALIGN_CENTRE) 
        box.Add(cblbl,0,wx.EXPAND|wx.ALIGN_CENTER_HORIZONTAL|wx.ALL,5)
        
        languages = ['C', 'C++', 'Python', 'Java', 'Perl'] 
        self.combo = wx.ComboBox(panel,choices = languages) 
        box.Add(self.combo,1,wx.EXPAND|wx.ALIGN_CENTER_HORIZONTAL|wx.ALL,5) 
        chlbl = wx.StaticText(panel,label = "Choice control",style = wx.ALIGN_CENTRE) 

        box.Add(chlbl,0,wx.EXPAND|wx.ALIGN_CENTER_HORIZONTAL|wx.ALL,5) 
        self.choice = wx.Choice(panel,choices = languages) 
        box.Add(self.choice,1,wx.EXPAND|wx.ALIGN_CENTER_HORIZONTAL|wx.ALL,5)
        box.AddStretchSpacer() 
        self.combo.Bind(wx.EVT_COMBOBOX, self.OnCombo) 
        self.choice.Bind(wx.EVT_CHOICE, self.OnChoice)

        panel.SetSizer(box) 
        self.Centre() 
        self.Show() 
    
    def OnCombo(self, event): 
        self.label.SetLabel("You selected"+self.combo.GetValue()+" from Combobox") 

    def OnChoice(self,event): 
        self.label.SetLabel("You selected "+ self.choice.GetString(self.choice.GetSelection())+" from Choice")
        
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.ComboBox & Choice Demo")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
```

Guage进度条控件

```python

import wx
import time

class Frame(wx.Frame):
    """Frame class."""

    def __init__(self, parent=None, id=-1, title='Title',
                 pos=wx.DefaultPosition, size=(400, 200)):
        """Create a Frame instance."""
        wx.Frame.__init__(self, parent, id, title, pos, size)
        self.InitUI()
    
    def InitUI(self):
        #创建panel
        self.count = 0 
        pnl = wx.Panel(self) 
        vbox = wx.BoxSizer(wx.VERTICAL)

        hbox1 = wx.BoxSizer(wx.HORIZONTAL) 
        hbox2 = wx.BoxSizer(wx.HORIZONTAL)

        self.gauge = wx.Gauge(pnl, range = 20, size = (250, 25), style =  wx.GA_HORIZONTAL) 
        self.btn1 = wx.Button(pnl, label = "Start") 
        self.Bind(wx.EVT_BUTTON, self.OnStart, self.btn1) 

        hbox1.Add(self.gauge, proportion = 1, flag = wx.ALIGN_CENTRE) 
        hbox2.Add(self.btn1, proportion = 1, flag = wx.RIGHT, border = 10) 

        vbox.Add((0, 30)) 
        vbox.Add(hbox1, flag = wx.ALIGN_CENTRE) 
        vbox.Add((0, 20)) 
        vbox.Add(hbox2, proportion = 1, flag = wx.ALIGN_CENTRE) 
        pnl.SetSizer(vbox) 

        self.SetSize((300, 200)) 
        self.Centre() 
        self.Show(True)
        
    def OnStart(self, e): 
        while True: 
            time.sleep(1); 
            self.count = self.count + 1 
            self.gauge.SetValue(self.count) 
            if self.count >= 20: 
                print("end") 
                return
            
class App(wx.App):
    """Application class."""

    def OnInit(self):
        self.frame = Frame(title="wx.Panel")#顶层窗口
        self.frame.Show()#激活框架窗口
        self.SetTopWindow(self.frame)
        return True

def main():
    app = App()
    app.MainLoop()

if __name__ == '__main__':
    main()
```
